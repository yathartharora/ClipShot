```python
from copy import deepcopy
import codecs
import rule
from sacremoses import MosesTokenizer, MosesDetokenizer
from conllu import parse
from rule import Question, AnswerSpan
import pattern
from stanza.utils.conll import CoNLL
import stanza
```

```python
def convert_to_conll(question,ans_conll,nlp):
  doc = nlp(question)
  dicts = doc.to_dict()
  for i in range(len(dicts[0])):
    try:
      del dicts[0][i]['feats']
    except KeyError:
      pass
    try:
      del dicts[0][i]['start_char']
    except KeyError:
      pass
    try:
      del dicts[0][i]['end_char']
    except KeyError:
      pass
  conll = CoNLL.convert_dict(dicts)
  with open('outputs.conllu','a') as f:
    for i in conll:
      for j in i:
        for k in j:
          f.write(k)
          f.write("   ")
        f.write("\n")
    f.write("\n")
    for i in ans_conll:
      for j in i:
        for k in j:
          f.write(k)
          f.write("   ")
    f.write("\n\n")



```

```python
nlp = stanza.Pipeline('en')
ans_dict=nlp('mask')
ans_dicts = ans_dict.to_dict()
try:
  del ans_dicts[0][0]['feats']
except KeyError:
  pass
try:
  del ans_dicts[0][0]['start_char']
except KeyError:
  pass
try:
  del ans_dicts[0][0]['end_char']
except KeyError:
  pass
ans_conll = CoNLL.convert_dict(ans_dicts)
```

```python
detokenizer = MosesDetokenizer()
```

```python
print('Parsing conllu file...')
with codecs.open('outputs.conllu', 'r') as f:
    conllu_file = parse(f.read())
```

```python
conllu_file
```

```python
# Creating dict
ids = range(int(len(conllu_file)/2))
examples = {}
count = 0
for i, s in enumerate(conllu_file):
    if i % 2 == 0:
      examples[ids[count]] = s
    else:
        examples[str(ids[count])+'_answer'] = s
        count +=1
```

```python
def qa2d(idx):
    q = Question(deepcopy(examples[idx]))
    print(q.question)
    if not q.isvalid:
        print("Question {} is not valid.".format(idx))
        return ''
    a = AnswerSpan(deepcopy(examples[str(idx)+'_answer']))
    if not a.isvalid:
        print("Answer span {} is not valid.".format(idx))
        return ''
    q.insert_answer_default(a)
    return detokenizer.detokenize(q.format_declr(), return_str=True)
```

```python
def print_sentence(idx):
    return detokenizer.detokenize([examples[idx][i]['form'] for i in range(len(examples[idx]))], return_str=True)
```

```python
total = int(len(examples.keys())/2)
count=0
import csv
# Specify the file name
file_name = "questions_templates_dp.csv"

print("Transforming {} examples.".format(total))
data=[["Question","Template"]]
for i in range(total):
    d=[]
    try:
      out = qa2d(i)
    except AttributeError:
      pass
    d.append(print_sentence(i))
    count+=1
    if out != '':
        d.append(out)
    data.append(d)
# Writing data to the CSV file
with open(file_name, mode='w', newline='') as file:
    writer = csv.writer(file)

    # Writing header
    writer.writerow(data[0])

    # Writing data rows
    writer.writerows(data[1:])
print("Total: ",count)
```
